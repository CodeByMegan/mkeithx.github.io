```python
def fib(n):

    a, b = 0, 1

    while a < n:
        print(a, end=' ')
        a, b = b, a+b
    
    print()

fib(2048)
```

# Plots and Maths

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set_theme(style="darkgrid")
```

```python
a = np.linspace(0,64,256)

plt.plot(a,np.cos(a))
plt.title("Cosine")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```
